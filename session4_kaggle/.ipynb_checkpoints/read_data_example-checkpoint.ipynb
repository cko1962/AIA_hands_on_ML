## How to make submission file
當訓練好模型後，使用 test.csv (只有 features 沒有 label 的資料) 進行預測，將得到的預測結果輸出成符合格式的 csv，並上傳至 Kaggle 的網頁來得到成績

```python
## test 是我們要預測的資料。 example 則是官方提供的上傳範例格式
test = pd.read_csv("advance_titanic_survivor_prediction/test.csv")
example = pd.read_csv("advance_titanic_survivor_prediction/gender_submission.csv")
```

```python
## testing data 的 features，注意此處沒有 Survived 的 label，因為這就是我們要預測的！
test.head()
```

```python
## 上傳結果必須要符合這樣的格式
example.head()
```

```python
example.tail()
```

```python
# 假設預測結果如下
import random
y_pred = np.array([0] * 418)
y_pred[random.sample(range(418), 209)] = 1
y_pred
```

```python
# 將預測結果做成 dataframe （表格形式）
my_submission = pd.DataFrame({"PassengerID": list(range(892, 1310)), "Survived": y_pred})
my_submission.head()
```

```python
# 將預測結果輸出成 csv，注意一定要加上 index=False，否則會多一個 column 叫做 index
my_submission.to_csv("random_submission.csv", index=False)
```
